```python
import os
import cv2
from ultralytics import YOLO

def crop_players_from_frames(frames_folder, output_crops_folder, model_path, conf_threshold=0.5, player_class=2):
    """
    Crop all players from every frame in frames_folder using YOLO model, and save crops.

    Args:
        frames_folder (str): Path to folder with extracted frames.
        output_crops_folder (str): Where to save cropped player images.
        model_path (str): YOLO model weights path.
        conf_threshold (float): Confidence threshold for detections.
        player_class (int): Class ID for players in the model.
    """
    os.makedirs(output_crops_folder, exist_ok=True)
    model = YOLO(model_path)

    frame_files = sorted([f for f in os.listdir(frames_folder) if f.endswith(('.jpg', '.png'))])
    print(f"Processing {len(frame_files)} frames...")

    total_crops = 0

    for frame_id, frame_file in enumerate(frame_files):
        frame_path = os.path.join(frames_folder, frame_file)
        frame = cv2.imread(frame_path)

        results = model(frame)[0]

        player_id = 0
        for box in results.boxes:
            cls = int(box.cls[0])
            conf = float(box.conf[0])
            if cls == player_class and conf > conf_threshold:
                x1, y1, x2, y2 = map(int, box.xyxy[0])
                crop = frame[y1:y2, x1:x2]
                crop_filename = f"{os.path.splitext(frame_file)[0]}_p{player_id:02d}.jpg"
                crop_path = os.path.join(output_crops_folder, crop_filename)
                cv2.imwrite(crop_path, crop)
                player_id += 1
                total_crops += 1

        if frame_id % 10 == 0:
            print(f"Processed {frame_id+1}/{len(frame_files)} frames...")

    print(f"Saved {total_crops} player crops in '{output_crops_folder}'")
```

```python
import torchreid

# Load model on CPU
model = torchreid.models.build_model(
    name='osnet_x0_25',
    num_classes=1000,
    pretrained=True
)
model.eval()  # No .cuda()

# Updated extract_features_from_folder (use CPU)
from torchvision import transforms
from PIL import Image
import torch
import os
import numpy as np

# Transform for input image
transform = transforms.Compose([
    transforms.Resize((256, 128)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406],
                         std=[0.229, 0.224, 0.225])
])

def extract_features_from_folder(folder, model):
    features_dict = {}
    for filename in sorted(os.listdir(folder)):
        if filename.endswith(".jpg"):
            path = os.path.join(folder, filename)
            image = Image.open(path).convert('RGB')
            tensor = transform(image).unsqueeze(0)  # Keep on CPU
            with torch.no_grad():
                feature = model(tensor)
            features_dict[filename] = feature.numpy().flatten()
    return features_dict
```

```python
features_broadcast = extract_features_from_folder("crops_broadcast", model)
features_tacticam = extract_features_from_folder("crops_tacticam", model)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

def match_players(features_tacticam, features_broadcast, threshold=0.75):
    matches = {}
    for t_name, t_feat in features_tacticam.items():
        best_match = None
        best_score = -1  # Cosine similarity ranges from -1 to 1
        for b_name, b_feat in features_broadcast.items():
            sim = cosine_similarity(
                t_feat.reshape(1, -1), b_feat.reshape(1, -1)
            )[0][0]
            if sim > best_score:
                best_score = sim
                best_match = b_name
        if best_score >= threshold:
            matches[t_name] = (best_match, best_score)
        else:
            matches[t_name] = (None, best_score)
    return matches
```

```python
matched_players = match_players(features_tacticam, features_broadcast, threshold=0.75)

# Let's inspect top 5 matches
for i, (t_name, (b_name, score)) in enumerate(matched_players.items()):
    print(f"{i+1}. Tacticam: {t_name} ↔ Broadcast: {b_name} | Score: {score:.3f}")
    if i >= 4:
        break
```

```python
# Assign fixed IDs to broadcast crops first
broadcast_id_map = {}
for i, b_name in enumerate(features_broadcast.keys()):
    player_id = f"P{str(i+1).zfill(3)}"
    broadcast_id_map[b_name] = player_id

# Now map each tacticam player to broadcast ID (if matched)
final_player_ids = {}

for t_name, (b_name, score) in matched_players.items():
    if b_name is not None:
        final_player_ids[t_name] = broadcast_id_map[b_name]
    else:
        # Assign new ID for unmatched players
        new_id = f"P{len(broadcast_id_map) + 1}"
        broadcast_id_map[t_name] = new_id
        final_player_ids[t_name] = new_id
```

```python
import torch
import torchreid
from torchvision import transforms
from PIL import Image
import os

# Load pretrained osnet_x1_0 model for person re-identification
model = torchreid.models.build_model(
    name='osnet_x1_0',
    num_classes=1000,
    pretrained=True
)
model.eval()
model = model.to('cuda' if torch.cuda.is_available() else 'cpu')

# Image transforms (resize, normalize etc.)
transform = transforms.Compose([
    transforms.Resize((256, 128)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406],
                         std=[0.229, 0.224, 0.225])
])

def extract_features_from_folder(folder_path, model):
    device = 'cuda' if torch.cuda.is_available() else 'cpu'
    features = {}
    for img_name in os.listdir(folder_path):
        img_path = os.path.join(folder_path, img_name)
        img = Image.open(img_path).convert('RGB')
        img_t = transform(img).unsqueeze(0).to(device)

        with torch.no_grad():
            feat = model(img_t)
            feat = feat.squeeze().cpu()
            feat = feat / feat.norm()  # Normalize the feature

        features[img_name] = feat
    return features

# Example usage:
# features_broadcast = extract_features_from_folder("crops_broadcast", model)
# features_tacticam = extract_features_from_folder("crops_tacticam", model)
```

```python
import torch
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np

def cosine_similarity(feat1, feat2):
    # Convert numpy arrays to torch tensors if needed
    if isinstance(feat1, np.ndarray):
        feat1 = torch.from_numpy(feat1)
    if isinstance(feat2, np.ndarray):
        feat2 = torch.from_numpy(feat2)
    return torch.nn.functional.cosine_similarity(feat1.unsqueeze(0), feat2.unsqueeze(0)).item()
def match_players(features_tacticam, features_broadcast, threshold=0.85):
    matches = []
    for tac_img, tac_feat in features_tacticam.items():
        best_match = None
        best_score = -1
        for brd_img, brd_feat in features_broadcast.items():
            score = cosine_similarity(tac_feat, brd_feat)
            if score > best_score:
                best_score = score
                best_match = brd_img

        if best_score >= threshold:
            matches.append((tac_img, best_match, best_score))
    return matches

# Example usage:
matches = match_players(features_tacticam, features_broadcast, threshold=0.90)
print(f"Found {len(matches)} matches with score >= 0.90")

# Visualize matches side-by-side with score
def visualize_matches(matches, folder_tacticam, folder_broadcast, max_display=5):
    plt.figure(figsize=(10, 4 * max_display))
    for i, (tac_img, brd_img, score) in enumerate(matches[:max_display]):
        tac_path = os.path.join(folder_tacticam, tac_img)
        brd_path = os.path.join(folder_broadcast, brd_img)

        tac_img_pil = Image.open(tac_path)
        brd_img_pil = Image.open(brd_path)

        plt.subplot(max_display, 2, 2*i + 1)
        plt.imshow(tac_img_pil)
        plt.title(f"Tacticam: {tac_img}")
        plt.axis('off')

        plt.subplot(max_display, 2, 2*i + 2)
        plt.imshow(brd_img_pil)
        plt.title(f"Broadcast: {brd_img}\nScore: {score:.3f}")
        plt.axis('off')
    plt.tight_layout()
    plt.show()

# Call to visualize top matches
visualize_matches(matches, 'crops_tacticam', 'crops_broadcast')
```

```python
import easyocr
import cv2
import numpy as np

# Initialize EasyOCR reader once (English only)
reader = easyocr.Reader(['en'])

def extract_jersey_number(image_path):
    # Read image
    img = cv2.imread(image_path)
    if img is None:
        return None
    # Convert to grayscale
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    # Crop region where jersey number is expected (e.g. center or back)
    # This depends on your dataset; for now, try whole image or bottom half
    h, w = gray.shape
    roi = gray[int(h*0.5):h, :]  # bottom half crop, adjust if needed

    # Run OCR on roi
    result = reader.readtext(roi)

    # Filter text results for numbers (jersey numbers usually 1-3 digits)
    numbers = []
    for (bbox, text, prob) in result:
        if text.isdigit() and 0.5 < prob < 1.0:  # threshold confidence
            numbers.append(text)

    # Return most confident detected number or None
    if numbers:
        return numbers[0]
    return None
```

```python
import cv2
from easyocr import Reader

reader = Reader(['en'])

def preprocess_for_ocr(img):
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    gray = cv2.equalizeHist(gray)
    _, thresh = cv2.threshold(gray, 150, 255, cv2.THRESH_BINARY_INV)
    return thresh

def extract_jersey_number(image_path):
    img = cv2.imread(image_path)
    if img is None:
        print(f"Failed to read image: {image_path}")
        return None

    h, w, _ = img.shape

    rois = [
        img[int(h*0.3):int(h*0.7), :],  # middle slice
        img[int(h*0.5):h, :],           # bottom half
        img,                           # full crop
    ]

    for i, roi in enumerate(rois):
        processed_roi = preprocess_for_ocr(roi)

        # For debugging, save processed ROI images
        cv2.imwrite(f"debug_roi_{i}.jpg", processed_roi)

        result = reader.readtext(processed_roi)
        numbers = []
        for (bbox, text, prob) in result:
            if text.isdigit() and prob > 0.5:
                numbers.append(text)
        if numbers:
            return numbers[0]
    return None
```

```python
import os

def extract_jersey_numbers_for_folder(folder_path):
    jersey_numbers = {}
    for filename in os.listdir(folder_path):
        if filename.endswith(".jpg") or filename.endswith(".png"):
            img_path = os.path.join(folder_path, filename)
            jersey_num = extract_jersey_number(img_path)
            jersey_numbers[filename] = jersey_num
            print(f"{filename} - Jersey Number: {jersey_num}")
    return jersey_numbers

# Run for both folders
jersey_broadcast = extract_jersey_numbers_for_folder("crops_broadcast")
jersey_tacticam = extract_jersey_numbers_for_folder("crops_tacticam")
```
